```python
# Model Building
 #it is difficult to visualize high dimensional data, we can use PCA to find the first two principal components

from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_pca = std_scaler.fit_transform(wine2)
scaled_pca
```

```python
#Check Collinearity : PCA will function better if the features are highly correlated.
wine2.corr().round(2)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(scaled_pca)
transformed_data
pca_data = pd.DataFrame(transformed_data)
pca_data.columns = ['PC1','PC2']
pca_data
```

```python
variance = pca.explained_variance_ratio_ #How much information each Principal component takes
variance
```

```python
import numpy as np
np.cumsum(np.round(a = variance,decimals=4)*100)
```

```python
plt.scatter(pca_data['PC1'], pca_data["PC2"], alpha=.5, c=wine_data['Type'])
plt.show()
```

**The scatter plot given an indication that there may be 3 clusters present**

#### Finding out the optimal number of clusters

```python
from sklearn.cluster import AgglomerativeClustering , KMeans
```

```python
#using elbow method
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i,random_state = 12)
    kmeans.fit(scaled_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The scree plot levels off at k=3 and let's use it to determine the clusters

**1. Hierarchial clustering**

```python
# create dendrogram
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(scaled_pca, method='ward'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward')
hc
```

```python
hc.fit_predict(scaled_pca)
```

```python
y_hc=hc.fit_predict(scaled_pca)
```

```python
Clusters_hc=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters_hc.head(5)
```

```python
wine2['clusters_hc'] = Clusters_hc
wine2
```
